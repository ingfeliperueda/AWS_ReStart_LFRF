```python
import subprocess

def install_or_remove_packages():
    iOrR = ""
    while iOrR not in ("I", "R"):
        print("Would you like to install or remove packages? (I/R)")
        iOrR = input().upper()

    action = "install" if iOrR == "I" else "remove"

    # Ejemplo: ejecutar apt, yum, pacman, etc.
    # Ajusta el comando según tu distro
    cmd = ["sudo", "apt", action, "package-name"]

    result = subprocess.run(cmd, capture_output=True, text=True)

    print("STDOUT:", result.stdout)
    print("STDERR:", result.stderr)
    print("Return code:", result.returncode)
```
